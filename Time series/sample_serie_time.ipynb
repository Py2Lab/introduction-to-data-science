```python
import pandas as pd
%matplotlib inline
import seaborn; seaborn.set()
import matplotlib.pyplot as plt
import numpy as np
```

### Visualizaci√≥n de recuentos de bicicletas de Seattle

```python
# Quitar comentario para descargar archivo 
#!curl -o FremontBridge.csv https://data.seattle.gov/api/views/65db-xm6k/rows.csv?accessType=DOWNLOAD
```

```python
data = pd.read_csv('FremontBridge.csv', index_col='Date', parse_dates=True)
data.head()
```

```python
data.columns = ['Total', 'West', 'East']
data['Total'] = data.eval('West + East')
```

```python
data.dropna().describe()
```

Visualizando los datos
Podemos obtener una idea del conjunto de datos al visualizarlo. Comencemos por graficar los datos sin procesar:


```python
data.plot()
plt.ylabel('Hourly Bicycle Count');
```

Las ~25,000 muestras por hora son demasiado densas para que podamos entenderlas. Podemos obtener m√°s informaci√≥n volviendo a muestrear los datos en una cuadr√≠cula m√°s gruesa. Remuestreemos por semana:


```python
weekly = data.resample('W').sum()
weekly.plot(style=[':', '--', '-'])
plt.ylabel('Weekly bicycle count');
```

Otra forma que resulta √∫til para agregar los datos es usar una media m√≥vil, utilizando la funci√≥n pd.rolling_mean(). Aqu√≠ haremos una media m√≥vil de 30 d√≠as de nuestros datos, asegur√°ndonos de centrar la ventana:

```python
daily = data.resample('D').sum()
daily.rolling(30, center=True).sum().plot(style=[':', '--', '-'])
plt.ylabel('mean hourly count');
```

La irregularidad del resultado se debe al corte duro de la ventana. Podemos obtener una versi√≥n m√°s suave de una media m√≥vil usando una funci√≥n de ventana, por ejemplo, una ventana gaussiana. El siguiente c√≥digo especifica tanto el ancho de la ventana (elegimos 50 d√≠as) como el ancho de la Gaussiana dentro de la ventana (elegimos 10 d√≠as):

```python
daily.rolling(50, center=True,
              win_type='gaussian').sum(std=10).plot(style=[':', '--', '-']);
```

### Profundizando en los datos

Si bien estas vistas de datos suavizados son √∫tiles para tener una idea de la tendencia general de los datos, ocultan gran parte de la estructura interesante. Por ejemplo, podr√≠amos querer ver el tr√°fico promedio como una funci√≥n de la hora del d√≠a.


```python
by_time = data.groupby(data.index.time).mean()
hourly_ticks = 4 * 60 * 60 * np.arange(6)
by_time.plot(xticks=hourly_ticks, style=[':', '--', '-']);
```

El tr√°fico horario tiene una distribuci√≥n fuertemente bimodal, con picos alrededor de las 8:00 de la ma√±ana y las 5:00 de la tarde. Esta es una evidencia probable de un fuerte componente de tr√°fico de pasajeros que cruza el puente. Esto se evidencia a√∫n m√°s por las diferencias entre la acera occidental (que generalmente se usa para ir hacia el centro de Seattle), que alcanza un pico m√°s fuerte en la ma√±ana, y la acera este (que generalmente se usa para alejarse del centro de Seattle), que tiene un pico m√°s fuerte en la noche.

Tambi√©n podemos tener curiosidad acerca de c√≥mo cambian las cosas seg√∫n el d√≠a de la semana. Nuevamente, podemos hacer esto con un simple groupby:

```python
by_weekday = data.groupby(data.index.dayofweek).mean()
by_weekday.index = ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun']
by_weekday.plot(style=[':', '--', '-']);
```
