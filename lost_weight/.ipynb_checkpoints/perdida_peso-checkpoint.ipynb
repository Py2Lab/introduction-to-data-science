```python
def plot_two_variables(label_x,label_y,data_frame):

    """We have 6 diferents groups, this function
    plot two variables"""

    fig, ax = plt.subplots(figsize=(10, 6))
    ax.scatter(x = data_frame[label_x], y = data_frame[label_y])
    plt.xlabel(label_x)
    plt.ylabel(label_y)
    plt.savefig('distribucion_perdida_edad.png')
    plt.show()
```

```python
plot_two_variables("Clase",
                    "Porcentaje_de_perdida",
                    df)
```
