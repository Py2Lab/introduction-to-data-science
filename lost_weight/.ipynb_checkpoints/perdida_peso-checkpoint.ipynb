```python
def fileter_class(number_class):
    class_filtered = df[df.Clase==number_class]
    return class_filtered

no_conclusion = fileter_class(0)
cancer = fileter_class(1)
disseminated_infection = fileter_class(2)
autoimmune_disease = fileter_class(3)
non_neoplastic_gastrointestinal = fileter_class(4)
psychiatric = fileter_class(5)
miscellaneous = fileter_class(6)
```

```python
def get_n_variables(dataframe, *argv):
    list_variables = [i for i in argv]
    return dataframe[list_variables]

get_n_variables(df, "IMC Previo","Porcentaje_de_perdida").corr()
```

```python
data_frame_lose_weight = get_n_variables(df,"Fecha Ingreso",
                                            "Fecha Egreso",
                                            "Costo por hospitalizacion"
                                            )
```

```python
def plot_two_variables(label_x,label_y,data_frame):

    """We have 6 diferents groups, this function
    plot two variables"""

    fig, ax = plt.subplots(figsize=(10, 6))
    ax.scatter(x = data_frame[label_x], y = data_frame[label_y])
    plt.xlabel(label_x)
    plt.ylabel(label_y)
    plt.savefig('distribucion_perdida_edad.png')
    plt.show()
```

```python
plot_two_variables("Fecha Ingreso",
                    "Fecha Egreso",
                    df)
```

```python
data_frame_lose_weight["DIAS"] = (data_frame_lose_weight["Fecha Egreso"]-data_frame_lose_weight["Fecha Ingreso"]).dt.days
```
